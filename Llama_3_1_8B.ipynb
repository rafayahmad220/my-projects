```python
import transformers
import torch
import os

model_id = "meta-llama/Meta-Llama-3.1-8B"

# Define the chat template
def format_chat_template(messages):
    sys_message = messages[0]["content"]
    user_message = messages[1]["content"]
    chat_input = f"<s>[INST] <<SYS>>\n{sys_message}\n<</SYS>>\n{user_message} [/INST]"
    return chat_input

# Initialize the pipeline
pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
    tokenizer=transformers.AutoTokenizer.from_pretrained(model_id)
)
```

```python
# Prepare messages
messages = [
    {"role": "system", "content": "You are a helpful, respectful and honest assistant."},
    {"role": "user", "content": "Which is the smallest country in the world?"}
]

# Format the chat input
chat_input = format_chat_template(messages)

# Generate the response
outputs = pipeline(chat_input, max_new_tokens=256)

# Print the generated text
print(outputs[0]["generated_text"])
```

```python
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import torch

# Define the model ID
model_id = "meta-llama/Meta-Llama-3.1-8B"

# Configure 4-bit quantization
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,               # Enable 4-bit loading
    bnb_4bit_use_double_quant=True,  # Use double quantization (optional, improves performance for large models)
    bnb_4bit_quant_type="nf4",       # Quantization type (use 'nf4' for better quantization)
    bnb_4bit_compute_dtype=torch.float16  # Set the compute dtype to FP16
)

# Load the model and tokenizer with disk offloading
with torch.inference_mode():
    model = AutoModelForCausalLM.from_pretrained(
        model_id,
        device_map="auto",                   # Let Transformers decide the best placement
        quantization_config=bnb_config,      # Use the BitsAndBytesConfig for 4-bit quantization
        offload_folder="offload",            # Specify a folder for offloading
        offload_state_dict=True              # Enable state dict offloading (helps reduce memory usage)
    )
    tokenizer = AutoTokenizer.from_pretrained(model_id)

# Initialize the text generation pipeline
text_gen = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
# Example function to evaluate MMLU (Massive Multitask Language Understanding) dataset
def evaluate_mmlu(mmlu_dataset):
    correct = 0
    total = len(mmlu_dataset['train'])
    for sample in mmlu_dataset['train']:
        question = sample['What is at risk for Lymphocytic Choriomeningitis (LCM)?']  # Adjust key based on dataset structure
        options = sample['options']    # Adjust if needed
        correct_answer = sample['LCMV infections can occur after exposure to fresh urine, droppings, saliva, or nesting materials from infected rodents. Transmission may also occur when these materials are directly introduced into broken skin, the nose, the eyes, or the mouth, or presumably, via the bite of an infected rodent. Person-to-person transmission has not been reported, with the exception of vertical transmission from infected mother to fetus, and rarely, through organ transplantation.']    # Adjust if needed

        input_text = f"What is at risk for Lymphocytic Choriomeningitis (LCM)? {question}\nOptions: {', '.join(options)}"

        # Generate the model's response
        output = text_gen(input_text, max_new_tokens=50)[0]['generated_text']

        # Extract the generated answer (you may need to adjust this depending on model output)
        generated_answer = output.split('\n')[-1].strip()  # Extracts the last line (likely the answer)

        print(f"Question: {question}")
        print(f"Model's answer: {generated_answer}")
        print(f"Correct answer: {correct_answer}")
        print("----")

        # Compare the generated answer with the correct answer
        if generated_answer == correct_answer:
            correct += 1

    accuracy = correct / total
    print(f"MMLU Accuracy: {accuracy * 100:.2f}%")
```
